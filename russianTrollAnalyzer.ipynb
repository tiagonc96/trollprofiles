```python
highestScore=[]

for i in trollUsers.index:
    score=has_normal_name(trollUsers['name'][i])+ get_numbers_in_profile(trollUsers['screen_name'][i],trollUsers['name'][i])+contains_keywords(trollUsers['id'][i])+is_account_created_after(trollUsers['created_at'][i])+get_following_to_follower_ratio(trollUsers['friends_count'][i],trollUsers['followers_count'][i])
    if(score>=3):
        highestScore.append(trollUsers['name'][i])
```

```python
def get_numbers_in_profile(troll,name):
    username= troll
    numberCount=0
    for letter in username:
        
        if(letter.isdigit()):
            numberCount+=1
            
    if(numberCount>=4):
        return 1
    else:
        try:
            for letter in name:
                if(str(letter).isdigit()):
                    numberCount+=1
            
                if(numberCount>=4):
                    return 1
        except:
            return 0

    return 0
```

```python
from names_dataset import NameDataset

def has_normal_name(troll):

    names = NameDataset()
    
    
    try:
        trollName= troll.split()
        if(names.search_first_name(trollName[0]) or names.search_last_name(trollName[1])):
            return 0
            
        if(not names.search_first_name(trollName[0]) or not names.search_last_name(trollName[1])):
            return 1
    
    except:
           return 0
    
    
```

```python
def contains_keywords(trollUser):
    found=0
    
    keywords=['trump','police','hillary','clinton','black','obama','state','american','woman','shooting','killed','donald','people','clinton','refugee','school','president','shot','terrorist','country','win','muslim','fire','officer','death']
    
    
    for index in trollTweets.index:
        if(trollTweets['user_id'][index]==trollUser):
           
            for keyword in keywords:        
                if(keyword in str((trollTweets['text'][index])).lower()):
                    return 1
    
   
    return 0
    
```

```python
def get_following_to_follower_ratio(following,followers):

    if(following-followers>200):
        return 1
    else:
        return 0
           
```

```python
def is_account_created_after(troll):

   
    try:
            fullDateArray=troll.split()
            shortDate= fullDateArray[2]+" "+fullDateArray[1]+" "+fullDateArray[5]
            d1= datetime.strptime(shortDate,'%d %b %Y')
            a= datetime(2014,3,15)
            if(d1 > a):
                print("datas a funcionar")
                return 1
            else: 
                return 0
    except:
        return 0
```
