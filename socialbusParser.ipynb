```python
import pandas as pd
import os
import json
import glob


#for filename in os.listdir("D:\\socialbus\\users\\192.168.1.75\\json\\2020\\02\\08\\"):
filenames=[]
jsons=[]
for filename in glob.iglob("D:\\socialbus\\users\\*" + '*', recursive=True):
    try:
        jsons.append(pd.read_json(filename,lines=True))
        print(filename)
        print("deu")
    except:
        #filenames.remove(filename)
        print(filename)
        print("nao deu")
        
```

```python
document= ""

for json in jsons:
    rows= len(json['text'])
    i=0

    while(rows>1):
            document = document +"\n"+json['text'][i]
            i+=1
            rows-=1
    
print(document)
```
