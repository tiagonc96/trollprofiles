```python
def get_numbers_in_profile(jsonLine):
    username= jsonLine['screen_name']
    numberCount=0
    for letter in username:
        if(letter in '[0-9]'):
            numberCount+=1
            
    if(numberCount>=4):
        return 1
    else:
        return 0
        
```

```python
def is_br(jsonLine):
    if('brasil' in str(jsonLine['location']).lower() or 'brazil' in str(jsonLine['location']).lower() or 'brasileir' in str(jsonLine['description']).lower() or 'brasil' in str(jsonLine['description']).lower() or 'flameng' in str(jsonLine['description']).lower()):
        return True
    else:
        return False
```

```python
def contains_keywords(jsonText,jsonUser):
    found=0
   
    
    keywords=['nacionalis','patriota','comuna','comunista','esquerdista','esquerda','esquerdalh','direita','fascista','facho','troll','chega','socialismo','liberal','preto',' be ',' il ',' ps ','ventura','livre','joacine','conservador','eutanasia','puta','caralho','merda','populista','populismo','europa']
    for word in keywords:
        textoTweet= str(jsonText).lower()
        textoBio= str(jsonUser['description']).lower()
        username=jsonUser['name'].lower()
        if word in textoTweet or word in textoBio or word in username:
            found = 1
            break
    return found
    
```
